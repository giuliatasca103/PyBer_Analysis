# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
ride_count_by_type = pyber_data_df.set_index(['type']).groupby(['type']).count()['ride_id']
ride_count_by_type
```

```python
# 2. Get the total drivers for each city type
driver_count_by_type = city_data_df.groupby(['type']).sum()['driver_count']
driver_count_by_type
```

```python
#  3. Get the total amount of fares for each city type
fare_count_by_type = pyber_data_df.set_index(['type']).groupby(['type']).sum()['fare']
fare_count_by_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride_type = fare_count_by_type / ride_count_by_type
avg_fare_per_ride_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver_by_type = fare_count_by_type / driver_count_by_type
avg_fare_per_driver_by_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(
                    {'Total Rides': ride_count_by_type,
                     'Total Drivers': driver_count_by_type,
                     'Total Fares': fare_count_by_type,
                     'Average Fare per Ride': avg_fare_per_ride_type,
                     'Average Fare per Driver': avg_fare_per_driver_by_type})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df = pd.DataFrame(
                    {'Total Rides': ride_count_by_type.map("{:,}".format),
                     'Total Drivers': driver_count_by_type.map("{:,}".format),
                     'Total Fares': fare_count_by_type.map("${:,.2f}".format),
                     'Average Fare per Ride': avg_fare_per_ride_type.map("${:,.2f}".format),
                     'Average Fare per Driver': avg_fare_per_driver_by_type.map("${:,.2f}".format)})
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Rename columns in combined ride dataframe
pyber_data_df.rename(columns = {'city': 'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'}, inplace=True)
pyber_data_df
```

```python
# Set the index to the data
pyber_data_df = pyber_data_df.set_index(['Date'])
```

```python
pyber_data_df.index = pd.to_datetime(pyber_data_df.index)
```

```python
# Make a copy with fewer columns
pyber_data_df_two = pyber_data_df.copy()[['Fare', 'City Type']]
pyber_data_df_two
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
# Create a new dateframe grouped by city type and date
pyber_data_df_fare_by_type = pd.DataFrame(pyber_data_df_two.groupby(['City Type', 'Date']).sum()['Fare'])
pyber_data_df_fare_by_type.reset_index()
pyber_data_df_fare_by_type
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_pivot = pd.pivot_table(pyber_data_df_fare_by_type, values="Fare", index=["Date"], columns=['City Type'])
pyber_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_pivot_df = pyber_pivot.loc['2019-01-01':'2019-04-29']
pyber_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pyber_pivot_df.index = pd.to_datetime(pyber_pivot_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_pivot_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_pivot_weekly_df = pyber_pivot_df.resample('W').sum()
pyber_pivot_weekly_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

# plot
ax = pyber_pivot_weekly_df.plot(figsize=(20, 7.5))

ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.set_xlabel("Month")


plt.savefig("analysis/Pyber_fare_summary.png")
plt.show()
```
